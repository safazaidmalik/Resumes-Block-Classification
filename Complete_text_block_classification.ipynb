```python
import re
import os
import string
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
stopwords = set(stopwords.words('english'))

def dataCleaning(filepath):
    file = open(filepath, 'r')
    # .lower() converts upper case characters to lower case characters.
    text = file.read().lower()
    file.close()
    # replaces anything that is not a lowercase letter, a space, or an apostrophe with a space:
    text = re.sub('[^a-z\ \']+', " ", text)
    words = list(text.split())
    non_stopwords = []
    for s in words:
        if filepath != "/home/szm/Desktop/KDD_Internship/final/annotated_files/":
            if s not in stopwords and len(s) != 1:
                non_stopwords.append(s)
        elif filepath == "/home/szm/Desktop/KDD_Internship/final/annotated_files/skills":
            if s not in stopwords:
                non_stopwords.append(s)
    return non_stopwords



def extractVocablury(C):  
    D = {}
    for i in  C:                  # For loop iterating each class
        if i not in D:   # If the specific word is not already in a dictionary
            D[i]=1                # Append new unique word.
    return D

def countTokens(vocab,doc):
    D = {}
    for w in vocab:  # Loop iterating all unique words of vocablury
        if w in doc:   # If word in vocablury also present in loop
            D[w] = doc[w]       # Store the number of instances stored above
        else:
          #
            D[w] = 0            # This condition ensures a error will not appear when iterating vocablury loop in main function to avoid key error
    return D

def concatenateTexts(c):
    D = {}
    for i in c:
        if i not in D:
            D[i] = 1      # If not in dict
        else:
            D[i]+=1       # Increment if word is already present.
    return D

def TrainMultinomialNB(C,classes):
    V = extractVocablury(set(C))
    N= 50 # len(T)
    cond_prob = dict()
    for i in V:
        cond_prob[i] = dict()

    prior={}
    for c in range(len(classes)):
        Nc = 50                       # Len of the training texts in class c
        Nw = len(classes[c])          # Words in all text of the class c
        prior[c] = 1                  # Storing Prior
        doc = concatenateTexts(classes[c])  # Containing all words of class c
        #print('hi',doc,'bye')
        Ni = countTokens(V,doc)       # Tokens of each word of vocablury in the class (loops over vocablry words in the function)
        for w in V.keys():
            cond_prob[w][c] = (Ni[w] + 1) / (Nw + len(V))   #Storing Conditional Prob for every word.

    return (V,prior,cond_prob)



# path = "/home/szm/Desktop/KDD_Internship/final/annotated_files/"
path = "/home/szm/Desktop/KDD_Internship/final/class_files/"
f_list = os.listdir(path)
writing_path = "/home/szm/Desktop/KDD_Internship/task5_files_cleaning/files/merged_list.txt"

# Create vocab by concatenating the 4 lists and creating set.
vocab = []
edu_list =  dataCleaning(path+"education")
vocab += edu_list
skl_list =  dataCleaning(path+"skills")
vocab += skl_list
exp_list =  dataCleaning(path+"experience")
vocab += exp_list
oth_list =  dataCleaning(path+"others")
vocab += oth_list


classes = {}
classes[0] = edu_list
classes[1] = skl_list
classes[2] = exp_list
classes[3] = oth_list
# print(len(classes[0]))
# print(len(classes[1]))
# print(len(classes[2]))
# print(len(classes[3]))

# print(len(classes[0])+len(classes[1])+len(classes[2])+len(classes[3]))


# classes[3]

y = TrainMultinomialNB((edu_list+skl_list+exp_list+oth_list),classes)
word_probs = y[2]
# print(y[2])


from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
stopwords = set(stopwords.words('english'))

def blockCleaning(block):
    block = block.lower()
    # replaces anything that is not a lowercase letter, a space, or an apostrophe with a space:
    block = re.sub('[^a-z\ \']+', " ", block)
    words = list(block.split())
    non_stopwords = []
    for s in words:
        if s not in stopwords:
            non_stopwords.append(s)
    return set(non_stopwords)


import math
import glob
import numpy as np
'''
Read separate blocks from a CV (later from all CVs) and find labels for each block
A block starts after the tag "###This is a block tag###"
-> Open file in read mode.
-> Create a string/list.
-> Keep reading file and appending to string/list till read line == "###This is a block tag###"
-> Remove the tag from string/list
-> Send to the next function:-
    * for a text block:-
        remove duplicates(place in a set)
        remove prepositions, stopwords, punctuation, numbers, and convert to lower
        prob = 0
        create a list with size 4:- (one value for each class)
        for each class[i]:-
            for each word in text block and in vocab:-
                prob[i]+= log(probability of word in list l[i])

        argmax = max(prob)

        * the index at which argmax lies is the class this text block belongs to.

'''
# creating list of strings. Each string is a block from the CVs.

list_blocks = []
out_txt_path = "/home/szm/Desktop/KDD_Internship/final/block_converted_CVs/"

blocks_list = []
# for file in glob.glob(out_txt_path+"/Creative Designer_Moazzam Baig.txt_out.txt"):
# out_txt_path = "/home/szm/Desktop/KDD_Internship/final/test_block_files/"
nb_classes_path = "/home/szm/Desktop/KDD_Internship/final/NB_classes/"

nb_f0 = open(nb_classes_path + "education.txt", 'w')
nb_f0 = open(nb_classes_path + "education.txt", 'a')
nb_f1 = open(nb_classes_path + "skills.txt", 'w')
nb_f1 = open(nb_classes_path + "skills.txt", 'a')
nb_f2 = open(nb_classes_path + "experience.txt", 'w')
nb_f2 = open(nb_classes_path + "experience.txt", 'a')
nb_f3 = open(nb_classes_path + "others.txt", 'w')
nb_f3 = open(nb_classes_path + "others.txt", 'a')

for file in glob.glob(out_txt_path+"/*_out.txt"):
    f = open(file, "r")
    line = f.read()
    blocks = (line.split("###This is a block tag###"))
    print(blocks)
    new_line = line.replace("\n", " ")
    new_line = line.replace(".", " ")
    new_blocks = (new_line.split("###This is a block tag###"))    
    print(new_blocks)

    for i in range(len((new_blocks))):
        blocks_list.append("")
        blocks_list[i] = blockCleaning(new_blocks[i])
        block_class_probs = [None] * 4
        
        for j in range(4):
            block_class_probs[j] = 0
        '''
        block_class_probs[0] = edu_list
        block_class_probs[1] = skl_list
        block_class_probs[2] = exp_list
        block_class_probs[3] = oth_list
        '''
        #about to classify the blocks
        prob =0
        for word in blocks_list[i]:
            if word in vocab:
                block_class_probs[0]+= math.log(word_probs[word][0])
                block_class_probs[1]+= math.log(word_probs[word][1])
                block_class_probs[2]+= math.log(word_probs[word][2])
                block_class_probs[3]+= math.log(word_probs[word][3])
        if np.argmax(block_class_probs) == 0:
            nb_f0.write(blocks[i])
        elif np.argmax(block_class_probs) == 1:
            nb_f1.write(blocks[i])
        elif np.argmax(block_class_probs) == 2:
            nb_f2.write(blocks[i])
        elif np.argmax(block_class_probs) == 3:
            nb_f3.write(blocks[i])
        else:
            print("No block written.")
            
        print(np.argmax(block_class_probs))
        
    print(blocks_list, sep = '\n'+'\n')

nb_f0.close() 
nb_f1.close()
nb_f2.close()
nb_f3.close()

# print(*blocks, sep='\n'+'\n')
```
